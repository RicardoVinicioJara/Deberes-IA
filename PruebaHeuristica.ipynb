    Consulta


```python
MATCH (start:Iglesia {name: 'I. Ministerial de Dios'}), (end:Iglesia {name: 'I. Nombre de Jesus'})
CALL gds.alpha.shortestPath.astar.stream({
  nodeQuery: 'MATCH (p:Iglesia) RETURN id(p) AS id',
  relationshipQuery: 'MATCH (p1:Iglesia)-[r:CONNECTION]->(p2:Iglesia) RETURN id(p1) AS source, id(p2) AS target, r.time AS weight',
  startNode: start,
  endNode: end,
  relationshipWeightProperty: 'weight',
  propertyKeyLat: 'latitude',
  propertyKeyLat: 'longitude'
})
YIELD nodeId, cost
RETURN gds.util.asNode(nodeId).name AS Iglesia, cost
```

![resolucion_ejercicio1](https://raw.githubusercontent.com/RicardoVinicioJara/IMG/master/nodosneo4jHC.png) 

```python
MATCH (start:Iglesia {name: "I. Ministerial de Dios"}), (end:Iglesia {name: "I. Nombre de Jesus"})
CALL gds.alpha.shortestPath.stream({
  nodeProjection: 'Iglesia',
  relationshipProjection: {
  CAMINO: { 
  type: 'CAMINO',
  properties: 'cost',
  orientation:'UNDIRECTED'
  } 
  },
  startNode: start,
  endNode: end,
  relationshipWeightProperty: 'cost'
  })
  YIELD nodeId, cost
  RETURN gds.util.asNode(nodeId).name AS NOMBRE, cost
```
